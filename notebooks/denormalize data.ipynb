```python
import sqlalchemy as sa
import pandas as pd
from datetime import datetime, timedelta
from pathlib import Path
```

```python
data_files = Path(r'../data').glob('*.csv')
df = pd.DataFrame()
for fn in data_files:
    with fn.open(encoding='Latin1') as f:
        df_ = pd.read_csv(f)
        df_.columns.values[-1] = 'year'
        df = df.append(df_)
```

```python
df['ongeldig'] = df[' ongeldige/blanco stemmen'].\
    add(df[' blanco stemmen'], fill_value=0).\
    add(df[' ongeldige stemmen'], fill_value=0)
df.drop([' ongeldige/blanco stemmen',' blanco stemmen', ' ongeldige stemmen'], axis=1, inplace=True)
```

```python
df.groupby('year')[' kiesgerechtigden'].sum()
```

```python
normal = pd.melt(df, id_vars=['CBS code', 'Amsterdamse code', 
                              'year',' kiesgerechtigden', 
                              ' geldige stemmen', 'ongeldig'])
normal = normal[normal['value'].notnull()]
```
